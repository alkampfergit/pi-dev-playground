# 00 — Hello TypeScript

The simplest possible check: can we run **TypeScript** in a Jupyter notebook? No dependencies, just `console.log`.

**Kernel:** select **Deno** (top-right). If you see output below the cells, the toolchain works. See `../README.md` for setup.

```typescript
console.log("Hello from TypeScript in a notebook! 🎉");
```

```typescript
// A tiny bit of real TypeScript: a typed function.
const greet = (name: string): string => `Hello, ${name}!`;
console.log(greet("pi.dev"));
```

```typescript
// Deno supports top-level await natively — handy for the pi.dev async API.
await new Promise((r) => setTimeout(r, 50));
console.log("top-level await works ✔");
```
